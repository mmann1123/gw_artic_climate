# examine data structure

```python
# open dataset
ds = xr.open_dataset('air.mon.mean.nc')
print(ds)
```

```python
# view variable called air
ds.air
```

```python
# select a single year of air temp data
ds.air.sel(time='1948-01-01')
```

```python
# select a time period of air data
ds.air.sel(time=slice("2000-06-01", "2000-06-10"))
```

```python
# look at time dimension
ds.time
```

```python
# see all years
ds.time.dt.year.head(20)
```

```python
# see all months
ds.time.dt.month.head(20)
```

```python
# select months jun, july
ds.time.sel(time=ds.time.dt.month.isin([6, 7])).head()
```

```python
# select months jun, july in two years
ds.time.sel(time=ds.time.dt.month.isin([6, 7]) & ds.time.dt.year.isin([1962,1973]))
```

```python
ds.time.dt.season
```

```python

ds_clim = ds.groupby('time.month').mean(dim='time')
ds_clim.air.sel(month=1).plot()

#%%
ds_anom = ds.groupby('time.month') - ds_clim
ds_anom.air[2].plot()
```
